```python
import numpy as np
import torch
import pandas as pd
from retriver import DenseRetrievalExactSearch
from sentence_transformers import SentenceTransformer
```

```python
dataset = pd.read_csv("../data/bgg_corpus.csv")# Please change the file name based on your data. 
corpus_size = len(dataset)
corpus = {}
for _, row in dataset.iterrows():
    ID = str(row['id']) 
    title = row['title']
    text = row['text']
    corpus[ID] = {
                "title": str(title).strip(),
                "text": str(text).strip()
            }
print("Corpus size: ", corpus_size)
```

```python
device = torch.device("mps" if torch.backends.mps.is_available() else "cpu")
model = SentenceTransformer(
    "intfloat/multilingual-e5-large-instruct",
    trust_remote_code=True
).to(device)

top_k = 5
score_function = "cos_sim"
batch_size=128
```

```python
queries = {
    "q1": "What is the best game for a family?",
    "q2": "What is the best game for a party?",
}
```

```python
# from google.colab import drive
# drive.mount('/content/drive')
# %cd /content/drive/MyDrive/Rice/COMP540/self_supervised
```

```python
dense_retriever = DenseRetrievalExactSearch(
    model, # Next page
    batch_size, 
    corpus_size,
    cache_dir="../cache"
)
```

```python
result = dense_retriever.search(
    corpus, # { doc_id: { title: text } }
    queries, # { query_id: text }
    top_k, 
    score_function, # One of {'cos_sim': cos_sim, 'dot': dot_score}
    return_sorted=True, # return sorted results
)

# { query_id: { doc_id: similarity (float)} }
```
